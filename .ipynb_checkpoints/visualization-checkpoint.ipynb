## Linear two class classification

The first dataset is a two class well seperated enough to be classified with a single line. The output of all neural network we train on this notebook will be a vector. Each element indicates the probabiliy of corresponding class, independent of others.

```python
x, y = generate_two_class_linear_dataset()
plot_dataset(x, y)
```

Before we test our first dataset on CapsuleNet, we want to see how a classic neural network behaves on above dataset.

```python
model = Sequential()
model.add(Dense(2, activation="sigmoid", input_shape=(2,)))
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

A two layer Neural network is like:

```python
model = Sequential()
model.add(Dense(2, activation="sigmoid", input_shape=(2,)))
model.add(Dense(2, activation="sigmoid"))
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

Now we try CapsuleNet:

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(2, 2, 1))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

As demonstrated in the above plot, although the capsule net is capable of solving non-linear problems, it was not able to seperate two class from each other. Now lets try moving one class to another location in which data is not symmetrical according to any axis.

```python
x, y = generate_two_class_linear_dataset_asym()
plot_dataset(x, y)
```

As it is shown, the two class are not symmetrical in any axis individually. Lets try CapsuleNet again.

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(2, 2, 1))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

The capsules seem to work better when the data is not symmetrical! But the truth is that since capsule layer uses matrix multiplication only, it can exploit the value of each dimension. On other hand in the prevoius example the data was actually spreaded through one dimention only, because the mean of both classes in $x_2$ dimension were equal to $0$ and the only thing that seperated them apart was the value of $x_1$ axis. So the entry data (one dimentional data) to capsule layer were not enough for two class classifcation task.

Now lets change value of dynamic routing iterations count and see how the results vary from routing iteration count of one(no dynamic routing).

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(2, 2, 2))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

As you can see, a little bit of curvature around the premiter of each cluster had happend. Now lets increase the routing iterations and compare the results.

3 Iteraions:

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(2, 2, 3))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

4 Iterations:

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(2, 2, 4))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

5 Iterations:

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(2, 2, 5))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

It is obviouse that the amount of non-linearity is intensified by increasing the routing iterations. It can help model to perform better on non-linear data but it may also leads to miss classification. So cross-validaiton for routing iterations is a must for each layer of capsules in the model.

## The XOR problem

We want to demonstrate that how CapsuleNet is able to solve the classic non-linear problem. But first lets generate our synthetic dataset and see what is the solution of a classic neural network for this problem is.

```python
x, y = generate_two_class_xor_dataset()
plot_dataset(x, y)
```

```python
model = Sequential()
model.add(Dense(2, activation="sigmoid", input_shape=(2,)))
model.compile("adam", mean_absolute_error)
model.fit(x, y, epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

As it was expected, the neural network is not able to fit to the data more than 50%. Now lets try to make the nework non-linear by adding more layers to it.

```python
model = Sequential()
model.add(Dense(2, activation="sigmoid", input_shape=(2,)))
model.add(Dense(2, activation="sigmoid"))
model.compile("adam", mean_absolute_error)
model.fit(x, y, epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

By adding more layers to classic neural network we were able to some how fit our model to our data. Now lets try it with 3 layers

```python
model = Sequential()
model.add(Dense(2, activation="sigmoid", input_shape=(2,)))
model.add(Dense(2, activation="sigmoid"))
model.add(Dense(2, activation="sigmoid"))
model.compile("adam", mean_absolute_error)
model.fit(x, y, epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

Added layer just somehow reduced the indecisiveness of model and didn't add any special capability to it. We want to try our dataset on CapsuleNet now.

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(2, 2, 1))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

The images speeks for themselfs! A one layer model with capsule did about %25 better than single layer classic neural network, But the images might raise a question in your brain that why wasn't the capsule layer able to classify the red dots completely currect the same way as it did for blue dots?
To answer this question, we will modify out ploting function to include a wider range.

```python
plot_decision_boundary(model.predict, x, y, lims=(-15, 25))
```

Now it is getting intersting. CapsuleNet founded the vector to perfectly seperate the red dots but somehow it was not able to move the vector to the currect position. The reason for that is "bios". Because CapsuleNet has no bios to add to multiplication result before activation, it was not able to solve the problem. To solve the problem we move our center of dataset(means of all data) to the center position(0, 0).

```python
x, y = generate_two_class_xor_centered_dataset()
plot_dataset(x, y)
```

Lets train CapsuleNet again.

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(2, 2, 1))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

On new dataset, CapsuleNet is able to achive %100 accuracy which is about 2 times more accurate than a classic neural network. From this experiment you can understand that if you are using CapsuleNet for any task you should somehow manage to prepare you data(pre-processing or by any hidden layer in model) to suit CapsuleNet. Let's increase the iterations of dynamic routing now.

Two iterations:

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(2, 2, 2))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

Three iterations:

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(2, 2, 3))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

Four iteration:

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(2, 2, 4))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

## What is really going on?

The CapsuleNet has done a greate job to solve the XOR problem. But the real problem is how is it doing this? We picked four arbitrarily point and check the output of capsule layer for each one and plot the output. Then we expected that we would be able to interpret what is CapsuleNet doing.

```python
x1 = np.array([[-5, -5]])
x2 = np.array([[5, 5]])
x3 = np.array([[-5, 5]])
x4 = np.array([[5, -5]])
```

Creating a function to do so:

```python
def test_single_point_on_xor_capsule(x):
    m = Model(inputs=model.inputs, outputs=[model.layers[1].output])
    y = m.predict(x).reshape((2, 2))
    x_ = x.reshape((2,))
    y_red = y[0, :]
    y_blue = y[1, :]
    print("Red point: ", y_red)
    print("Blue point: ", y_blue)
    plt.cla()
    plt.plot([0, 0], [-10, 10], "k-", lw=0.5)
    plt.plot([-10, 10], [0, 0], "k-", lw=0.5)
    plt.scatter(x_[0], x_[1], c="black")
    plt.scatter(y_red[0], y_red[1], c="red")
    plt.scatter(y_blue[0], y_blue[1], c="blue")
```

$X_1$:

```python
test_single_point_on_xor_capsule(x1)
```

$X_2$:

```python
test_single_point_on_xor_capsule(x2)
```

$X_3$:

```python
test_single_point_on_xor_capsule(x3)
```

$X_4$:

```python
test_single_point_on_xor_capsule(x4)
```

CapsuleNet managed to find two independent transformations that maps the correct class to a vector with length of nearly one and wrong class to a vector with length of nearly zero.

We can also inspect the learned weights in CapsuleNet to understand its behavior:

```python
model.layers[1].weights[0].eval(session=K.get_session()).reshape((2, 2, 2))
```

## Multi class classification task

We want to test CapsuleNet for multi class classification task. But first we need to generate our multi class classification dataset.

```python
x, y = generate_three_class_linear()
plot_three_class_dataset(x, y)
```

Lets try CapsuleNet:

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(3, 2, 1))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_three_class_decision_boundary(model.predict, x, y)
```

model is able to seperate three classes from each other. But lets increase the dynamic routing iterations in order to give model some non-linear decesion lines.

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(3, 2, 5))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_three_class_decision_boundary(model.predict, x, y)
```

With 5 iterations of dynamic routing the model was able to clearly distinguish different classes from each other.

## How well can dynamic routing help non-linearity

In this section we want to investigate the ability of dynamic routing in helping non-learity in a non-linear problem. Our synthetic datasets in prevoius sections was not indeed non-linear and any linear classifier was able to fit to that data. We continue our investigation by creating a non-linear dataset.

```python
from sklearn.datasets import make_moons
x, y = make_moons(400, shuffle=False, noise=0.1)
y = np.zeros((400, 2))
y[:200, 0] = 1
y[200:, 1] = 1
plot_dataset(x, y)
```

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(2, 2, 1))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

The pictures are chattering. CapsuleNet is not able to seperate the two moons from each other. Now we try with more iterations of dynamic routing.

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(2, 2, 3))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

Although the dynamic routing iterations make the situation a little better, our model needs something else to successfully seperate two moons from each other.

Lets try two layers of capsules:

```python
model = Sequential()
model.add(Lambda(lambda x: K.expand_dims(x), input_shape=(2,)))
model.add(CapsuleLayer(2, 2, 3))
model.add(CapsuleLayer(2, 2, 3))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

Two layers of capsules didn't help to fit to data that much, we need to try another architecture.

```python
model = Sequential()
model.add(Dense(25, input_shape=(2,)))
model.add(Reshape((5, 5)))
model.add(Activation(squash))
model.add(CapsuleLayer(2, 2, 1))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```

In above example we did make the model more accurate by combining the classic neural network with CapsuleNet. Classic fully-connected layer extracts latent features from input and capsule layer classify.

By combining classic neural network with CapsuleNet we got the moon curved line for seperation that we expected. The result for more iterations of dynamic routing is like:

```python
model = Sequential()
model.add(Dense(25, input_shape=(2,)))
model.add(Reshape((5, 5)))
model.add(Activation(squash))
model.add(CapsuleLayer(2, 2, 5))
model.add(Length())
model.compile("adam", mean_squared_error)
model.fit([x], [y], epochs=256, verbose=0)
plot_decision_boundary(model.predict, x, y)
```
